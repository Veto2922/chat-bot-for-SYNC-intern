## processing data

```python
words = [] #to stor word for tokenize
labels = [] #to stoe tags 
docs_x = []  #to stor normal word without tokenization
docs_y = []  #to store relationship between tags and words
```

```python
nltk.download('punkt')
```

```python
for intent in data ["intents"]:
        for pattern in intent["patterns"]:
            wrds = nltk.word_tokenize(pattern) #Separate each word in the sentence and put it in a list
            words.extend(wrds)
            docs_x.append(wrds)
            docs_y.append(intent["tag"])

        if intent["tag"] not in labels:
            labels.append(intent["tag"])
```

```python
stemmer = LancasterStemmer()
words = [stemmer.stem(w.lower()) for w in words if w != "?"] #Return the word to its origin and remove the question marks
words = sorted(list(set(words))) #function returns a sorted list of the specified iterable object
labels = sorted(labels)

```

```python
# words
```

```python
# labels
```

```python
training = []
output = []

out_empty = [0 for i in range(len(labels))] #list of zeros based on len of labels

for x, doc in enumerate(docs_x):
    bag_of_words = []

    wrds = [stemmer.stem(w) for w in doc]

    for w in words:  #Check if the words in one word list are also in the other list if yes = 1 else = 0
        if w in wrds:
            bag_of_words.append(1) # 
        else:
            bag_of_words.append(0)


    output_row = out_empty[:]
    output_row[labels.index(docs_y[x])] = 1

    training.append(bag_of_words)
    output.append(output_row)
```

```python
# output
```

```python
# training
```

```python
training = np.array(training)
output = np.array(output)
```

```python
#creat pickle file
with open("data.pickle", "wb") as f:
        pickle.dump((words, labels, training, output), f)
```

## Creat NN Model

```python
tf.reset_default_graph() 

net = tflearn.input_data(shape=[None, len(training[0])])
net = tflearn.fully_connected(net, 8)
net = tflearn.fully_connected(net, 8)
net = tflearn.fully_connected(net, len(output[0]), activation = "softmax")
net = tflearn.regression(net)

model = tflearn.DNN(net)
```

```python
try:
    model.load("model.tflearn")
except:
    model.fit(training, output, n_epoch=1000, batch_size=8, show_metric=True)
    model.save("model.tflearn")
```

```python
def bag_of_words(s, words): # processing input from user
    bag = [0 for _ in range(len(words))]

    s_words = nltk.word_tokenize(s)
    s_words = [stemmer.stem(word.lower()) for word in s_words]

    for se in s_words:
        for i, w in enumerate(words):
            if w == se:
                bag[i] = 1
            
    return numpy.array(bag)
```
